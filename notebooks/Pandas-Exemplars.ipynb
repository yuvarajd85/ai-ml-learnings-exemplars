```python
import numpy as np
from matplotlib import pyplot as plt
# Set up the figure
plt.figure(figsize=(8, 6))

# Original supply and demand curves
prices = np.linspace(2, 8, 100)  # Price levels
original_demand = 10 - 0.5 * prices  # Downward-sloping demand curve
original_supply = 0.5 * prices - 1  # Upward-sloping supply curve

# New supply and demand curves after changes
new_demand = 10 - 0.5 * prices  # Demand remains the same
new_supply = 0.5 * prices - 3  # Shift in supply curve (leftward due to reduced supply)

# Plot original curves
plt.plot(original_demand, prices, label="Original Demand", linestyle="--", color="blue")
plt.plot(original_supply, prices, label="Original Supply", linestyle="--", color="green")

# Plot new curves
plt.plot(new_demand, prices, label="New Demand", color="blue")
plt.plot(new_supply, prices, label="New Supply", color="green")

# Highlight equilibrium points
plt.scatter(6, 4, color="black", label="Original Equilibrium", zorder=5)  # Original equilibrium
plt.scatter(4, 5, color="red", label="New Equilibrium", zorder=5)  # New equilibrium

# Add labels and legend
plt.title("Supply-Demand Dynamics for Arabica Beans", fontsize=14)
plt.xlabel("Quantity", fontsize=12)
plt.ylabel("Price", fontsize=12)
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.legend(fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()

# Show the graph
plt.show()
```

```python
# Data for supply curve
quantity_supplied = np.linspace(0, 10, 100)  # Quantities supplied
marginal_cost = 2 + 0.5 * quantity_supplied**2  # Rising marginal cost

# Plot the supply curve
plt.figure(figsize=(8, 6))
plt.plot(quantity_supplied, marginal_cost, label="Supply Curve (Upward Sloping)", color="green")

# Add labels, title, and legend
plt.title("Upward-Sloping Supply Curve", fontsize=14)
plt.xlabel("Quantity Supplied", fontsize=12)
plt.ylabel("Price (Marginal Cost)", fontsize=12)
plt.legend(fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()

# Show the graph
plt.show()
```
