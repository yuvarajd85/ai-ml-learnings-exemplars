```python
#Two Sum Function - time complexity O(n) solution.
from typing import List
def two_sum(coll:List, target:int):
    temp_dict = {}
    for i,n in enumerate(coll):
        ele = target - n
        print(ele)
        if ele in temp_dict.keys():
            return (temp_dict[ele],i)
        else:
            temp_dict[n] = i
    return (-1,-1)
```

```python
data_coll = [2,3,5,9,11]
target = 8
print(two_sum(data_coll, target))
```
