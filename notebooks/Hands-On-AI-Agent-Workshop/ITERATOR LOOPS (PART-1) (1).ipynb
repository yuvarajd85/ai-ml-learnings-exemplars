# ----------------------------STORE VALUE LOOP-------------------------

```python
l=[1,2,3,1,5,7,4,11,33,21,2,4,6,8]

sums=0
for i in l:
    sums+=i
print(sums)
```

```python

l=[1,.5,1,2.5,1.5,1,2,3]

vessel=0
for i in l:
    vessel=vessel+i
print(vessel)
    
   
    
```

```python
l=[1,.5,1,2.5,1.5,1,2,3]
# variable to store the sum
vessel_even=0
vessel_odd=0
for i in l:
    if i%2==0:
        vessel_even=vessel_even+i
        print(vessel_even,"vessel_even")
    else:
        vessel_odd=vessel_odd+i
        print(vessel_odd,"vessel_odd")
       
```
