```python
my_pets = ['alfred', 'tabitha', 'william', 'arla']

result = []
def upper_case(num):
    
    for i in num:
        a = i.upper()
        result.append(a)
    return(result)

b=upper_case(my_pets)
b

```

```python
my_strings = ['a', 'b', 'c', 'd', 'e']
my_numbers = [1,2,3,4,5]

results = list(map(lambda x, y: (x, y), my_strings, my_numbers))
list(results)
```

#______________FILTER______________

Syntax

    filter(function, iterable)
    
filter() method returns an iterator that passed the function check for each element in the iterable.

```python



scores = [66, 90, 68, 59, 76, 60, 88, 74, 81, 65]

def is_student(score):
    return score > 75

over_75 = list(filter(is_student, scores))
over_75



```

```python
def even_odd(num):
    if num%2==0:
        return True

lf=[1,2,3,4,5,6,7,8]
list(filter(even_odd,lf))

```
