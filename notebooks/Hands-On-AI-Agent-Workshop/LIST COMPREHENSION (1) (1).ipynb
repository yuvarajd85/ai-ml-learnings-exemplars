```python
_______________________________________Nested IF with List COMPREHENSION_______________________________________________

Here, list comprehension checks:

Is y divisible by 2 or not?
Is y divisible by 7 or not?
If y satisfies both conditions, y is appended to num_list.
```

```python
df.loc[(df["salary"]<100000) & (df["gender"]=="female")]
```

```python




num_list = [y for y in range(100) if y % 2 == 0 if y % 7 == 0]
num_list

```
