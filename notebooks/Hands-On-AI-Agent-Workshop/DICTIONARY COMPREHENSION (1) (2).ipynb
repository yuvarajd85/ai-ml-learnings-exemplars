```python


original_dict = {'jack': 38, 'michael': 48, 'guido': 57, 'john': 33}

even_dict = {k: v for (k, v) in original_dict.items() if v %2==0}
even_dict
```

```python


____________________Multiple if Conditional Dictionary Comprehension___________________________
```

```python


original_dict = {'jack': 38, 'michael': 48, 'guido': 57, 'john': 33}

new_dict = {k: v for (k, v) in original_dict.items() if v % 2 != 0 if v < 40}
new_dict








```

```python
______________________________________________if-else Conditional Dictionary Comprehension__________________________
```

```python
original_dict = {'jack': 38, 'michael': 48, 'guido': 57, 'john': 33}

new_dict = {k: (v,'old' if v > 40 else 'young') for (k, v) in original_dict.items()}

new_dict
```

```python


nest_dic = { k1: {k2: k1 * k2 for k2 in range(1, 6)} for k1 in range(2, 5)}
nest_dic

--------------------------------------------------------------------------------------------------------------------

a = dict()
for k1 in range(2,5):
    a[k1] = dict()
    for k2 in range(1, 6):
        a[k1][k2] = k1*k2
print(a)
```
